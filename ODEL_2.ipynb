```python
import requests, openpyxl
import lxml
from bs4 import BeautifulSoup
from xlwt import *


excel = openpyxl.Workbook()
print(excel.sheetnames)
sheet = excel.active
sheet.title = 'Product Details'
print(excel.sheetnames)
sheet.append(['Product Number', 'Product Name', 'New Price', 'Old Price', 'discount'])


url = "https://glomark.lk/deal-page"
f= requests.get(url)

soup = BeautifulSoup(f.content, "lxml")

wb =Workbook(encoding = "utf-8")
ws = wb.add_sheet("dataset")


ws.write(0, 0, "Product number")
ws.write(0, 1, "Product name")
ws.write(0, 2, "New price")
ws.write(0, 3, "Old price")
ws.write(0, 4, "Discount")

details = soup.find_all("div", class_ = "col-6 col-sm-4 col-md-3 col-lg-2 col-xl-2")
print(details)

product_number = 0
line = 0

for i in details:
  product_number += 1
  line += 1

  item_name = i.find("span", class_ = "light-font").text.strip()
  item_discount = i.find("span").text.strip()
  new_price = i.find("div", class_="price").text.strip()
  old_price = i.find("div", class_="price").text.strip()

  print("item name:", item_name)
  print("new price:", new_price)
  print("old price:", old_price)
  print("item discount:", item_discount)
  
  ws.write(line, 0, product_number)
  ws.write(line, 1, item_name)
  ws.write(line, 2, new_price)
  ws.write(line, 3, old_price)
  ws.write(line, 4, item_discount)


excel.save('Glomark Product Details.xlsx')
```
